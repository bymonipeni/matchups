```python
import pandas as pd
```

```python
#download datasets
dfe = pd.read_csv("https://s3.amazonaws.com/csvpastebin/uploads/f9b678bf4a60bafe3f0a4132313e194d/event_interview.csv")
```

```python
dfo = pd.read_csv("https://s3.amazonaws.com/csvpastebin/uploads/e593dd089bc5aaf6583b5187f96cb48e/outcome_interview.csv")
```

```python
dff = pd.read_csv("https://s3.amazonaws.com/csvpastebin/uploads/4f1b611a0d9f96d7700adba8b387df21/fighter_interview.csv")
```

```python
dfe.head()
```

```python
dfo.head()
```

```python
#rename the column id from event dataset like in outcome dataset
dfe.rename(columns={'id':'event_id'}, inplace= True)
```

```python
dfe.head()
```

```python
#join event and outcome datasets
dfm= pd.merge(dfo, dfe, on='event_id', how='outer')
```

```python
dfm.head(10)
```

```python
# rename the column name for event_name 
dfm.rename(columns={'name':'event_name'}, inplace= True)
```

```python
dfm.head()
```

```python
# delete columns that we don't need
dfm1= dfm.drop(['id','winner_id','event_id','date','promoter','location','attendance','created_at','modified_at'],axis=1)
```

```python
# rename column id in fighter dataset
dff.rename(columns={'id':'fighter1_id'}, inplace= True)
```

```python
dff.head()
```

```python
# join the new dataset with the fighter dataset 
dfmfighter1= pd.merge(dfm1, dff, on='fighter1_id', how='outer')
```

```python
dfmfighter1.head()
```

```python
# rename columns like the clients wants
dfmfighter1.rename(columns={'nationality':'fighter1_nationality', 'name':'fighter1_name'}, inplace= True)
```

```python
dfmfighter1.head()
```

```python
# delete columns that we don't need
dfmfighter=dfmfighter1.drop(['fighter1_id'],axis=1)
```

```python
# rename column fighter1_id in fighter dataset for fighter2_id
dff.rename(columns={'fighter1_id':'fighter2_id'}, inplace= True)
```

```python
dff.head()
```

```python
# join the dataset with the fighter2_id  dataset
dfmfighter2= pd.merge(dfmfighter, dff, on='fighter2_id', how='outer')
```

```python
dfmfighter2.head()
```

```python
#rename columns like the client wants 
dfmfighter2.rename(columns={'nationality':'fighter2_nationality', 'name':'fighter2_name'}, inplace= True)
```

```python
dfmfighter2.head()
```

```python
# delete column that we don't need
dfmfighter2.drop(['fighter2_id'],axis=1)
```

```python
# order columns like the client wants
df_matchups = dfmfighter2[['fighter1_name','fighter2_name', 'fighter1_nationality', 'fighter2_nationality', 'event_name']]
```

```python
df_matchups.head()
```

```python
df_matchups.to_csv('matchups.csv')
```
